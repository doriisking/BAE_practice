# 문제
- 1. 'subscribed'컬럼이 2024년 2월인 데이터 찾기 
- 2. 1중 f3컬럼이 gold인 데이터 개수 구하기

# 오답
- datetime으로 바꿔야함 -> pd.to_datetime

```python
import pandas as pd
df = pd.read_csv("../data/type1_data1.csv")
```

```python
df.info()
```

```python
#1. 
df['subscribed'] = pd.to_datetime(df['subscribed'])
df.info()
```

```python
fev = df[(df['subscribed'].dt.year == 2024 )& (df['subscribed'].dt.month == 2 )]
fev
```

```python
#2. 
gold = fev[fev['f3']=='gold']
gold
```

```python
answer = len(gold)
answer
```
