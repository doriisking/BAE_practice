## 문제
1. 데이터에서 결측치 있는 행 모두 제거
2. 1에서부터 70%데이터 구하기 (70% 지점의 index가 소수점인 경우 소수점 이하 버림)
3. 2에서 views의 IQR정수로 구하기

## 오답
- 소수점버리기 -> import math 해서 math.floor(데이터)
- I3 사분위수 -> quantile(.75)
- I1 사분위수 -> quantile(.25)


```python
import pandas as pd
df = pd.read_csv("../data/type1_data1.csv")
```

```python
df.head()
```

```python
#1.
df = df.dropna()
```

```python
#2. 
import math
n = len(df) # 70
n_70  = math.floor(n*0.7)
n_70
```

```python
data_70 = df[:n_70]
data_70.shape
```

```python
#3. 
I3 = data_70['views'].quantile(.75)
I1 = data_70['views'].quantile(.25)
int(I3-I1)
```
