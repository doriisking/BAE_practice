# 오답
- "지역 **별**" 평균
   - groupby("지역")['열']
- groupby한 데이터에 대한 정렬
   - sort_values()
   - 그것에 대한 지역명은 index로.

```python
import pandas as pd
df = pd.read_csv("../data/type1_data1.csv")
df
```

```python
#1.
df_drop = df.dropna()
df_drop
```

```python
#2.
city_mean = df_drop.groupby('city')['f2'].mean()
city_mean
```

```python
#3. 
city_mean.sort_values(ascending=False).index[0]
```
