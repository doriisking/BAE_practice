## 문제 
- 1. 결측치가 많은 두 컬럼 찾기
- 2. 첫번째 많은 컬럼 : 결측치 행 삭제
- 3. 두번째 많은 컬럼 : 최빈값 대체
- 4. f3컬럼의 gold값가진 데이터 수 정수형 구하기


## 오답
- 특정컬럼 결측치 있을경우 행제거
: dropna(subset=[열])

```python
import pandas as pd
df = pd.read_csv("../data/type1_data1.csv")
```

```python
#1.
df.isnull().sum().sort_values(ascending=False)
first_null = df.isnull().sum().sort_values(ascending=False).index[0]
second_null = df.isnull().sum().sort_values(ascending=False).index[1]
first_null
```

```python
#2.
df= df.dropna(subset= first_null)
```

```python
#3.
m = df[second_null].value_counts().index[0]
df[second_null].fillna(m, inplace =True)
```

```python
#4.
len(df[df['f3'] == 'gold'])
```
