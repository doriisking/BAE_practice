2. 제일많은 컬럼에서 결측치 있는 데이터 삭제
3. 두번째로 많은 컬럼을 최빈값 대체
4. f3컬럼의 gold값 가진 데이터 수 정수형으로 구하기

## 오답
- 특정컬럼에 결측치가 있는경우 행제거

- 내 답
df = df[~df[first_null].isnull()] # true만 남기기

- 책 답
df = df.dropna(subset=['f1'])

```python
import pandas as pd
df = pd.read_csv("../data/type1_data1.csv")
```

```python
#1. 
a = df.isnull().sum().sort_values(ascending=False)
a
```

```python
first_null = a.index[0]
second_null = a.index[1]
print(first_null)
```

```python
#2.
df = df[~df[first_null].isnull()]
df.shape
```

```python
#3.
bigval = df[second_null].value_counts().index[0]
df[second_null].fillna(bigval,inplace = True)
```

```python
#4.
len(df[df['f3']=='gold'])
```
