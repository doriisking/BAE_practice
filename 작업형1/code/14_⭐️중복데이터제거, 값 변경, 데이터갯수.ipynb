## 문제
- 1. 중복데이터 제거
- 2. f3결측치 0, silver 1, gold2, vip 3 변환
- 3. f3컬럼 총합 -> 정수형

## 오답
- drop_duplicates()
- map 

```python
import pandas as pd
df = pd.read_csv("../data/type1_data1.csv")
```

```python
#1.
df.drop_duplicates(inplace=True)
```

```python
#2.
import numpy as np
df['f3'] = df['f3'].map({np.nan : 0, 'silver':1,'gold':2,"vip":3})
df['f3'].unique()
```
