## 문제

- 1. 주어진데이터에서 문자 자료형 컬럼 삭제
- 2. 숫자 자료형 컬럼의 결측치 0 대체
- 3. 각 행의 합이 3000 보다 큰 값의 개수를 정수로 구하기

## 오답
- select_dtypes (exclude!!!!!!)
- 각 행의 합 -> sum(axis=1)

```python
import pandas as pd
df = pd.read_csv("../data/type1_data1.csv")
```

```python
df.info()
```

```python
#1.
col = df.select_dtypes(exclude = 'object').columns
drop_ob = df[col]
```

```python
drop_ob
```

```python
#2.
drop_ob.fillna(0)
```

```python
#3.
count = drop_ob[drop_ob.sum(axis=1) > 3000]
```

```python
len(count)
```
