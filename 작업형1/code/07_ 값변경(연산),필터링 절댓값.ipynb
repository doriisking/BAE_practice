# 문제
- 1. 모든 나이(age)에 1을 더하기.
- 2. 20대 views 평균과 30대 views 평균 차이의 절댓값 구하기 (반올림 후 소수 둘째자리 까지 계산)

# 오답
- 절댓값은 abs(변수)

```python
import pandas as pd
df = pd.read_csv("../data/type1_data1.csv")
```

```python
#1.
df['age'] += 1 

df
```

```python
#2. 
twenty= df[(df['age'] >= 20) & (df['age'] <30)]
             
views20 = twenty['views'].mean()
```

```python
views20
```

```python
thirty = df[(df['age'] >= 30) & (df['age'] <40)]
             
views30 = thirty['views'].mean()
```

```python
views30
```

```python
answer = views20-views30
answer = abs(round(answer,2))
answer
```
