```python
# scaling 적용할 수치형 컬럼명 리스트로 만들기
num_cols = ['age','fnlwgt','education.num','capital.gain','capital.loss','hours.per.week']
#num_cols = num_cols.drop('id')
num_cols
```

```python
def get_data():
    train_copy = train.copy()
    test_copy = test.copy()
    return train_copy, test_copy
```

```python
train_copy, test_copy = get_data()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
display(train_copy[num_cols].head())
```

## data split

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(train, y_train, test_size=0.2, random_state=42)
```

## 모델 학습, 평가

모델 학습

```python
from sklearn.ensemble import RandomForestClassifier

rf= RandomForestClassifier(random_state=0)
rf.fit(X_train ,y_train)
pred = rf.predict_proba(X_val)
```

```python
print(rf.classes_)
pred[:10] ## 확률이 반환
```

- predict_proba 는 각 클래스 확률을 다 예측해줌( 이진분류니까 두개값이 생기는 셈) => ROC_AUC
- predict는 예측된 클래스만 반환 => 정확도, F1 score, 정밀도, 재현율

평가

```python
c_pred = rf.predict(X_val)
from sklearn.metrics import f1_score
f1 = f1_score(y_val,c_pred, pos_label = ">50K")
f1
```
